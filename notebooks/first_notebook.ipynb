## 1.1. dataframe league by league

```python
def get_league_number(db,league):
    ''' 
    Input : the cursor to the database and the country name of the league wanted
    output the league ID
    '''
    query = f'''
        SELECT *
    FROM League l 
    '''
    db.execute(query)
    rows = db.fetchall()
    ## getting the ID of the league/country
    count = 0
    league  = league.lower()
    league = league[0].upper() + league[1:]
    for item in np.array(rows)[:,2]:
        if league in item:
            return np.array(rows)[count,1]
        else:
            count += 1
    print('This league doesnt exist')
    return None
    
```

```python
get_league_number(db,'spain')
```

```python
def list_league(db,league):
    # return the number of directors contained in the database
    query = '''
    SELECT id, country_id, league_id,season, stage, Match.date, match_api_id, home_team_api_id, away_team_api_id, home_team_goal, away_team_goal, home_player_1, home_player_2, home_player_3, home_player_4, home_player_5, home_player_6, home_player_7, home_player_8, home_player_9, home_player_10, home_player_11, away_player_1, away_player_2, away_player_3, away_player_4, away_player_5, away_player_6, away_player_7, away_player_8,away_player_9, away_player_10, away_player_11, B365H, B365D, B365A 
    FROM Match
    WHERE Match.league_id = ?
    AND Match.B365H IS NOT NULL
    AND Match.B365D IS NOT NULL
    AND Match.B365A IS NOT NULL
    '''
    league_id = str(get_league_number(db,league))
    db.execute(query,(league_id))
    rows = db.fetchall()
    return rows
```
